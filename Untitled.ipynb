```python
### Handle Outliers
### Calculate Outlier percentage
def outlier_percentage (col_name, df=df) :
  q1 = df[col_name].quantile(.25)
  q3 = df[col_name].quantile(.75)
  iqr = q3 - q1
  upper_bound = q3 + 1.5*iqr
  lower_bound = q1 - 1.5*iqr
  total = df[df[col_name] > upper_bound].shape[0]+df[df[col_name]  < lower_bound].shape[0]
  print(total/df.shape[0])
```

```python
outlier_percentage('bmi')
```

```python
def outlier_drop (col_name, df=df) :
  q1 = df[col_name].quantile(.25)
  q3 = df[col_name].quantile(.75)
  iqr = q3 - q1
  upper_bound = q3 + 1.5*iqr
  lower_bound = q1 - 1.5*iqr
  new_df  = df[(df[col_name] <= upper_bound) & (df[col_name] >= lower_bound)]
  return new_df
```

```python
### Drop outliers
df = outlier_drop('bmi')
```

```python
### Check for Colleration
corr = df.select_dtypes(include='number').corr()
sns.heatmap(corr, annot=True)
```

```python
### split data ---> Encoding data
x = df.drop('charges', axis = 1 )
y = df.charges
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.2)
```

```python
df.describe(include="object")
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(drop='first')
ohe.fit(x_train[['sex','smoker','region']])
```

```python
temp = ohe.transform(x_train[['sex','smoker','region']]).toarray()
x_train.drop(['sex','smoker','region'], axis=1, inplace=True)
x_train[['sex_male', 'smoker_yes', 'region_northwest', 'region_southeast',
       'region_southwest']] = temp
```

```python
temp = ohe.transform(x_test[['sex','smoker','region']]).toarray()
x_test.drop(['sex','smoker','region'], axis=1, inplace=True)
x_test[['sex_male', 'smoker_yes', 'region_northwest', 'region_southeast',
       'region_southwest']] = temp
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error
pred = lr.predict(x_train)
mean_absolute_error(y_train, pred)
```

```python
pred = lr.predict(x_test)
mean_absolute_error(y_test, pred)
```
